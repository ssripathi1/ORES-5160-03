```python
#identifying duplicate rows
duplicates = data[data.duplicated()]
```

```python
# Dropping intended columns
columns_to_drop = ['CCN', 'Address', 'City', 'State', 'Zip Code', 'Infection Performance Score Capped', 'Mortality Adjustment','Performance Month']
data.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
data.columns
```

```python
# Renaming column names
column_names = {
    'Total Resident Weeks': 'Resident Weeks',
    'Total Covid Infections': 'Covid Infections',
    'Infection Performance Score': 'Infection Score',
    'Facility Infection Rate  Per 1000 Resident Weeks': 'Facility Infection Rate',
    'County Infection Rate  Per 1000 Resident Weeks': 'County Infection Rate'
}

data.rename(columns=column_names, inplace=True)
```

```python
data.columns
```

```python
data.info()
```

```python
# Changing the columns data type
data['Resident Weeks'] = data['Resident Weeks'].astype('float')
data['Covid Infections'] = data['Covid Infections'].astype('int')
data['Final Payment'] = data['Final Payment'].astype('int')
```

```python
data.info()
```

```python
data.info()
```

```python
# Check for missing values
missing_values = data.isnull()
missing_values_count = missing_values.sum()
```

```python
# Creating infection rate which is a new column
data['Infection Rate'] = (data['Covid Infections'] / data['Resident Weeks']) * 1000
# Creating a new column Rate Ratio
data['Rate Ratio'] = data['Facility Infection Rate'] / data['County Infection Rate']
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data['Resident Weeks'])
plt.title('Resident Weeks')
```

```python
# Z score to identify and deal outliners
# find the limits
upper_limit = data['Resident Weeks'].mean() + 3*data['Resident Weeks'].std()
lower_limit = data['Resident Weeks'].mean() - 3*data['Resident Weeks'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
# find the outliers
data.loc[(data['Resident Weeks'] > upper_limit) | (data['Resident Weeks'] < lower_limit)]
```

```python
# capping - change the outlier values to upper (or) lower limit values
new_data = data.copy()
new_data.loc[(new_data['Resident Weeks']>=upper_limit), 'Resident Weeks'] = upper_limit
new_data.loc[(new_data['Resident Weeks']<=lower_limit), 'Resident Weeks'] = lower_limit
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(new_data['Resident Weeks'])
plt.title('Resident Weeks')
```
